# Pandas Data Cleaning

Data may not come in a usable state all the time. 

Here are some problems that might require that we clean the data first

1. missing column names 

2. missing / null values in the rows

3. wrong data types 

4. clearly false or incorrect data entries 

5. excess columns, where data in the column is not needed
